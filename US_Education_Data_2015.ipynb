```python
import pandas as pd
from matplotlib import pyplot as plt
#import the data
df1 = pd.read_csv('https://raw.githubusercontent.com/IlmaU/US-Education-Data/main/states_all.csv')

#remove columns that won't be needed
df1.drop(columns = ['PRIMARY_KEY','ENROLL','FEDERAL_REVENUE','STATE_REVENUE','LOCAL_REVENUE','SUPPORT_SERVICES_EXPENDITURE','OTHER_EXPENDITURE','CAPITAL_OUTLAY_EXPENDITURE'], inplace=True)


#renaming some columns
df1 = df1.rename(columns={"GRADES_PK_G": " No. of students in PK","GRADES_KG_G": "No. of students in KG","GRADES_4_G": "No. of Students in 4th Grade","GRADES_8_G": "No. of students in 8","GRADES_12_G": "No. of students in 12", "GRADES_1_8_G": "No. of students in 1_8", "GRADES_9_12_G": "No. of students in 9_12", "GRADES_ALL_G":"No. of students in all grades"})

#create a figure for avg reading score for 4th grades of each state in 2015
fig1 = plt.figure(figsize=[38,10])
ax1 = fig1.add_subplot()

#all 2015 data minus the extra info
data_2015 = df1.loc[(df1['YEAR']==2015)]
data_2015 = data_2015[data_2015['STATE']!='NATIONAL']
data_2015 = data_2015[data_2015['STATE']!='DODEA']
data_2015 = data_2015[data_2015['STATE']!='DISTRICT_OF_COLUMBIA']

#reading scores of 4th graders in 2015
read_2015 = data_2015['AVG_READING_4_SCORE']
#plotting a bar graph
ax1.bar(data_2015['STATE'],read_2015)
#displaying a certain range on y axis to better show the difference in scores
plt.ylim(200, 260)
#labelling the axes
ax1.set_xlabel('States', fontsize = 18)
ax1.set_ylabel('Avg Reading scores of 4th graders in 2015', fontsize = 18)
#rotating the x ticklabels so they can be easily read
ax1.set_xticklabels(data_2015['STATE'],rotation = 45,fontsize=14)
ax1.set_yticklabels(read_2015, fontsize=14)
plt.show()
sorted = data_2015.sort_values(by=['AVG_READING_4_SCORE'],ascending=False)
print('5 States with the highest 4th grade reading scores')
print(sorted[['STATE','AVG_READING_4_SCORE']].head(5))

#stst


#create a fig for avg math score for 4th grades for each state in 2015
fig1 = plt.figure(figsize=[38,10])
ax1 = fig1.add_subplot()
#math scores of 4th graders in 2015
math_2015 = data_2015['AVG_MATH_4_SCORE']
#plotting a bar graph
ax1.bar(data_2015['STATE'],math_2015)
#displaying a certain range on y axis to better show the difference in scores
plt.ylim(200, 260)
#labelling the axes 
ax1.set_xlabel('States', fontsize = 18)
ax1.set_ylabel('Avg Math scores of 4th graders in 2015', fontsize = 18)
#rotating the x ticklabels so they can be easily read
ax1.set_xticklabels(data_2015['STATE'],rotation = 45,fontsize=14)
ax1.set_yticklabels(math_2015, fontsize=14)
plt.show()
sorted = data_2015.sort_values(by=['AVG_MATH_4_SCORE'],ascending=False)
print('5 States with the highest 4th grade math scores')
print(sorted[['STATE','AVG_MATH_4_SCORE']].head(5))



#No.of students enrolled in 4th grade in 2015 in each State
fig1 = plt.figure(figsize=[38,10])
ax1 = fig1.add_subplot()
enroll_2015 = data_2015['No. of Students in 4th Grade']
ax1.bar(data_2015['STATE'],enroll_2015)
ax1.set_xlabel('States', fontsize = 18)
ax1.set_ylabel('No. of Students enrolled in 4th grade in 2015', fontsize = 18)
ax1.set_xticklabels(data_2015['STATE'],rotation = 45,fontsize=14)
ax1.set_yticklabels(enroll_2015, fontsize=14)
plt.show()
sorted = data_2015.sort_values(by=['No. of Students in 4th Grade'],ascending=False)
print('5 States with the most students enrolled in 4th grade')
print(sorted[['STATE','No. of Students in 4th Grade']].head(5))



#Instruction expenditure in 2015 of each State
fig1 = plt.figure(figsize=[38,10])
ax1 = fig1.add_subplot()
expend_2015 = data_2015['INSTRUCTION_EXPENDITURE']
ax1.bar(data_2015['STATE'],expend_2015)
ax1.set_xlabel('States', fontsize = 18)
ax1.set_ylabel('Instruction Expenditure in 2015 ($)', fontsize = 18)
ax1.set_xticklabels(data_2015['STATE'],rotation = 45,fontsize=14)
ax1.set_yticklabels(expend_2015, fontsize=14)
plt.show()
sorted = data_2015.sort_values(by=['INSTRUCTION_EXPENDITURE'],ascending=False)
print('5 States with the most expenditure on instruction')
print(sorted[['STATE','INSTRUCTION_EXPENDITURE']].head(5))


#Make a column of Ratio of instruction expenditure per students enrolled
df1['Ratio'] = df1['INSTRUCTION_EXPENDITURE']/df1['No. of students in all grades']

#plotting the ratio of instruction expenditure per students enrolled of each state in 2015
fig1 = plt.figure(figsize=[38,10])
ax1 = fig1.add_subplot()

data_2015 = df1.loc[(df1['YEAR']==2015)]
data_2015 = data_2015[data_2015['STATE']!='NATIONAL']
data_2015 = data_2015[data_2015['STATE']!='DODEA']
data_2015 = data_2015[data_2015['STATE']!='DISTRICT_OF_COLUMBIA']

ax1.bar(data_2015['STATE'],data_2015['Ratio'])
ax1.set_xlabel('States', fontsize = 18)
ax1.set_ylabel('Instruction Expenditure Per Total Students Enrolled In 2015', fontsize = 18)
ax1.set_xticklabels(data_2015['STATE'],rotation = 45,fontsize=14)
ax1.set_yticklabels(data_2015['Ratio'], fontsize=14)
#display only upto two decimal points on the y axis 
from matplotlib.ticker import FormatStrFormatter
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
plt.show()

#Shows the 5 states that spend the most on instruction expenditure/students enrolled
sorted = data_2015.sort_values(by=['Ratio'],ascending=False)
print('5 States with the most instruction expenditure per students enrolled')
print(sorted[['STATE','Ratio']].head(5))
print('\n')
#Shows the 5 states that spend the least on instruction expenditure/students enrolled
sorted = data_2015.sort_values(by=['Ratio'],ascending=True)
print('5 States with the least instruction expenditure per students enrolled')
print(sorted[['STATE','Ratio']].head(5))








```
